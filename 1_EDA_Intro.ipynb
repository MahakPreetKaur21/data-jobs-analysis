# Exploratory Data Analysis 
## Goal
Investigate top-paying roles and skills in the data science industry.
Use Python to explore a real-live dataset on job postings.
For job-seekers: use these insights to help find the best job opportunities..
## Questions to Answer
1. What are the most demanded skills for the top 3 most popular data roles?
2. How are in-demand skills trending for Data Analysts?
3. How well do jobs and skills pay for Data Analysts?
4. What is the most optimal skill to learn for Data Analysts? (High Demand AND High Paying)

```python
#importing libraries
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r'C:\Users\mahak\Downloads\data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_plot=df['job_title_short'].value_counts().to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='job_title_short',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('Number of Jobs per Job Title')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

### Countries to explore


```python
df_plot=df['job_country'].value_counts().to_frame().head(20)
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_country', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Country')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

## Companies to Explore

```python
df_plot=df['company_name'].value_counts().to_frame().head(20)
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per company')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

## Job Opertunities

```python
counts_wfh=df['job_work_from_home'].value_counts()
plt.pie(counts_wfh,labels=counts_wfh.index, autopct='%1.1f%%', startangle=90)
plt.title("Work from Home Offered")
plt.show()
# Degree Requirement
counts_degree = df['job_no_degree_mention'].value_counts()
plt.pie(counts_degree, labels=counts_degree.index, autopct='%1.1f%%',startangle=90)
plt.title("Degree Requirement")
plt.show()

# Health Insurance
counts_insurance = df['job_health_insurance'].value_counts()
plt.pie(counts_insurance, labels=counts_insurance.index, autopct='%1.1f%%', startangle=90)
plt.title("Health Insurance Offered")
plt.show()
```

## Exploratory Data Analysis for Data Analysts in India


```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## Filter for India Data Analyst roles

```python
df_DA_Ind=df[(df['job_country'] == 'India') & (df['job_title_short'] == 'Data Analyst')]
```

## Locations to explore


```python
df_plot=df_DA_Ind['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Job Locations for Data Analyst in India')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

## Job Opertunities

```python
counts_wfh=df_DA_Ind['job_work_from_home'].value_counts()
plt.pie(counts_wfh,labels=counts_wfh.index, autopct='%1.1f%%', startangle=90)
plt.title("Work from Home Offered")
plt.show()
# Degree Requirement
counts_degree = df_DA_Ind['job_no_degree_mention'].value_counts()
plt.pie(counts_degree, labels=counts_degree.index, autopct='%1.1f%%',startangle=90)
plt.title("Degree Requirement")
plt.show()

# Health Insurance
counts_insurance = df_DA_Ind['job_health_insurance'].value_counts()
plt.pie(counts_insurance, labels=counts_insurance.index, autopct='%1.1f%%', startangle=90)
plt.title("Health Insurance Offered")
plt.show()
```

## Companies To Explore

```python
df_plot=df_DA_Ind['company_name'].value_counts().to_frame().head(20)
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per company')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
